**Overriding (Ä°ptal Etme)**
Miras aldÄ±ÄÄ±mÄ±z sÄ±nÄ±fta mevcut bir metodu kendi sÄ±nÄ±fÄ±mÄ±zda da tanÄ±mlarsak metodu Ã§aÄÄ±rdÄ±ÄÄ±mÄ±zda artÄ±k kendi sÄ±nÄ±fÄ±mÄ±zdaki metod Ã§alÄ±ÅacaktÄ±r. Bu kavrama nesne tabanlÄ± programlamada methodu override etmek denir.

```python
class Personel():
    def __init__(self,sicil,ad,soyad,maas,birim):
        print("Personel sÄ±nÄ±fÄ± init Ã§aÄ±lÄ±ÅtÄ±")
        self.sicil = sicil
        self.ad = ad
        self.soyad = soyad
        self.maas = maas
        self.birim = birim
        
    def bilgiler(self):
        print("Sicil: {}, Ad: {}, Soyad: {}, Maas: {}, Birim:{}".format(self.sicil,self.ad, self.soyad,self.maas,self.birim))
    
    def zamYap(self,tutar):
        self.maas += tutar #self.maas = self.maas
        
    def birimDegis(self,yBr):
        self.birim= yBr 
        
class Yazilimci(Personel):
    #Yazilimci sÄ±nÄ±fÄ±nda da __init__ fonk tanÄ±mlayarak miras aldÄ±ÄÄ±mÄ±z sÄ±nÄ±fÄ±n(Personel) init metodunu iptal ettik.
    #ArtÄ±k yazÄ±lÄ±mcÄ± sÄ±nÄ±fÄ±ndan bir nesne tÃ¼rettiÄimizde kendi init fonksiyonumuzu Ã§alÄ±ÅtÄ±rdÄ±ÄÄ±mÄ±zÄ± 
    #print ettiÄimiz "YazÄ±lmci init fonk" mesajÄ±ndan anlÄ±yoruz.
    def __init__(self,sicil,ad,soyad,maas,birim,pDil):
        print("YazÄ±lmci init fonk")
        self.sicil = sicil
        self.ad = ad
        self.soyad = soyad
        self.maas = maas
        self.birim = birim
        self.pDil = pDil
    def pDilEkle(self,pDil):
        self.pDil.append(pDil)
    def pDilListele(self):
        for i in self.pDil:
            print(i)
```

```python
yazilimci1 = Yazilimci("123","Ä°smail","AYDIN",100,"BL",["c#","Python","Js"])
yazilimci1.bilgiler()
```

```python
yazilimci1.pDilEkle("NodeJS")
yazilimci1.zamYap(5000)
yazilimci1.bilgiler()
yazilimci1.pDilListele()
```

**Super Anahtar Kelimesi**

Miras aldÄ±ÄÄ±mÄ±z metodlarÄ± kendi metodlarÄ±mÄ±z iÃ§erisinde kullanabilmemizi saÄlar.

```python
class Personel():
    def __init__(self,sicil,ad,soyad,maas,birim):
        print("Personel sÄ±nÄ±fÄ± init Ã§aÄ±lÄ±ÅtÄ±")
        self.sicil = sicil
        self.ad = ad
        self.soyad = soyad
        self.maas = maas
        self.birim = birim
        
    def bilgiler(self):
        print("Sicil: {}, Ad: {}, Soyad: {}, Maas: {}, Birim:{}".format(self.sicil,self.ad, self.soyad,self.maas,self.birim))
    
    def zamYap(self,tutar):
        self.maas += tutar #self.maas = self.maas
        
    def birimDegis(self,yBr):
        self.birim= yBr 
        
class Yazilimci(Personel):
    def __init__(self,sicil,ad,soyad,maas,birim,pDil):
        print("YazÄ±lmci init fonk")
        super().__init__(sicil,ad,soyad,maas,birim)
        self.pDil = pDil
    def pDilEkle(self,pDil):
        self.pDil.append(pDil)
    def pDilListele(self):
        for i in self.pDil:
            print(i)
```

```python
yazilimci1 = Yazilimci("123","Ä°smail","AYDIN",100,"BL",["c#","Python","Js"])
yazilimci1.zamYap(5000)
yazilimci1.pDilEkle("NodeJS")
yazilimci1.bilgiler()
yazilimci1.pDilListele()
```
